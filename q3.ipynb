```python
import numpy as np
from matplotlib import pyplot as plt

N_A = 1.2e17
N_D = 1.0e17
S = 2.0e-4
T = 300
kB_T_over_e = 0.02586
n_i = 1.08e10

tau_p = (7.8e-13*N_D + 1.8e-31*N_D**2)**(-1)
tau_n = (3.45e-12*N_A + 9.5e-32*N_A**2)**(-1)

mu_p = 130 + 370/(1+(N_D/8e17)**1.25)
mu_n = 232 + 1180/(1+(N_A/8e16)**0.9)

e = 1.6021e-19
I_s = e*n_i**2*S*np.sqrt(kB_T_over_e)*(np.sqrt(mu_n/tau_n)*1/N_A + np.sqrt(mu_p/tau_p)*1/N_D)

def I(V):
    return I_s*(np.exp(V/kB_T_over_e) - 1)
```

```python
V = np.linspace(-1,1,1000)
fig = plt.figure(figsize=(12,9))

plt.xlabel('V[V]', fontsize='30')
plt.ylabel('I[A]', fontsize='30')
plt.grid()
plt.plot(V, I(V), linewidth='4')
plt.title('pnホモ接合理想ダイオードの電流-電圧特性', fontname='MS Gothic', fontsize='30')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.savefig('./diod-I-V-curve.png')
```
